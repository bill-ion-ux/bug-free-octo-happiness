```python
#Create two random numpy array of size n x p and p x m and then compute the dot product
n = 5
m = 4
p = 3
array_1 = np.random.randint(1,100,(n,p))
array_2 = np.random.randint(1,100,(p,m))
dot_product = np.dot(array_1,array_2) #compute the dot product
print(f"array 1 is \n{array_1}")
print(f"array 2 is \n{array_2}")
print(f"dot product is \n{dot_product}")
```
